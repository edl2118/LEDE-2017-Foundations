## Grouping with resample, not with groupby

Hmmm, looks like something bad might have happened to the housing industry at some point. Maybe we want to see some numbers instead of a graph? To do aggregate statistics on time series in pandas we use a method called `.resample()`, and we're going to tell it **to group the data by year.**

When we tell it to group by year, we need to give it a **special code**. I always get mine from this StackOverflow post http://stackoverflow.com/a/17001474 because it's much more convenient than the pandas documentation.

```python
#one major benefit of creating a datetime column is so we can zoom out of our data (resample) or zoom in (groupby)
#all the other columns go away because they are strings and you can't take the mean of strings
df.resample("A").mean()
```

```python
df.resample("A").mean().plot(y="val")
```

```python
#does the same thing
#i prefer this syntax, personally it makes the most sense to me 
df.resample("A")["val"].mean().plot()
```

```python
#still does the same thing
df["val"].resample("A").mean().plot()
```

```python
#what if we resample every three years
df.resample("3A").mean().plot(y="val")
```

```python
#what if you want to resample for every quarter
df.resample("Q").mean().plot(y="val")
```

Notice that it's **December of every year**. That still looks like too much data, though. What if we back out to every decade?

```python
df.resample("10A").mean().plot(y="val", figsize=(20,5))
```

Cool, right?

## Cyclical data (actually using groupby)

### What were the top 5 worst months?

```python
df.sort_values(by="val").head()
```

```python
#another way to get the same result
#both results only give us the month per year, it doesn't help us see aggregate monthly trends
df.groupby('per_name')['val'].sum().sort_values().head()
```

It seems like there might be a cycle ever year. Maybe houses are sold in the summer and not the winter? To do this we can't use resample - it's for putting time into buckets - we need to **group by the month.**

### Getting the month

We can't ask for the index column as "year" any more, but we can just use `df.index` instead.

```python
#How do we look more closely at the seasonality?
#Group the data by month -- we can use groupby and our datetime index by month
```

To get the month of each date, it's simply `df.index.month`.

```python
df.groupby(df.index.month).mean().sort_values("val")
```

```python
#What if we want to look at the pattern by day of the week
df.groupby(df.index.dayofweek).mean()
```

### Doing the groupby to view data by month

So when we do our groupby, we'll say **hey, we made the groups for you already**. Then we ask for the median number of houses sold.

```python
df.groupby(df.index.month).median()
```

```python
df.groupby(df.index.month).median().sort_values(by="val")
```

### Plot the results

```python
#What does the pattern look like by month?
df.groupby(df.index.month).median().plot(y="val")
```

# More details

You can also use **max** and **min** and all of your other aggregate friends with `.resample`. For example, what's the **largest number of houses sold in a given year?**

```python
df.resample("A")["val"].max().sort_values(ascending=False)
```

How about the fewest?

```python
df.resample("A")["val"].min().sort_values()
```

Since we're feeling crazy, why don't we plot the **average, max and min for each year?**

```python
df.resample("A")["val"].agg(["mean", "min", "max"]).plot(figsize=(20,5))
```

```python
#if you want to change the colors
df.resample("A")["val"].mean().plot(color="green", figsize=(15,5))
df.resample("A")["val"].max().plot(color="blue", figsize=(15,5))
df.resample("A")["val"].min().plot(color="red", figsize=(15,5))
```

## Using `.fill_between` to draw between lines

Getting maximums and minimums for years is a very common way of dealing with resampled time series. A fun graph to make is the `fill_between` graph, which colors between two areas.

### Step 1: Calculate the maxes and mins and get a list of x values

`ax.fill_between` requires three things

    - a list of x values (the dates),
    - a list of minimum values for those x values,
    - and a list of maximum values for those x values

```python
x_values = df.resample("A").index
```

```python
y_max = df.resample("A")["val"].max()
```

```python
y_min = df.resample("A")["val"].min()
```

### Step 2: Build the graph

### Step 3: Add the fill

`ax.fill_between` requires three things

    - a list of x values
    - a list of minimum values
    - and a list of maximum values

We have all of those, so let's draw it!

```python
ax = df.resample("A")["val"].mean().plot(color = "red", figsize=(15,5))
ax.fill_between(x_values, y_max, y_min, alpha=0.5)
```
