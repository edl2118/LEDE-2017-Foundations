## Which borough has the highest number of dogs per-capita?

You’ll need to merge in `population_boro.csv`

```python
ls
```

```python
boro_df = pd.read_csv("boro_population.csv")
```

```python
boro_df
```

```python
df_combined.head(1)
```

```python
final_df = df_combined.merge(boro_df, left_on='borough', right_on='borough')
```

```python
dog_pop_df = final_df['borough'].value_counts().reset_index()
```

```python
dog_pop_df['dog_pop'] = dog_pop_df['borough']
```

```python
del dog_pop_df['borough']
```

```python
dog_pop_df
```

```python
percapita_df = boro_df.merge(dog_pop_df, left_on='borough', right_on='index')
```

```python
percapita_df
```

```python
percapita_df['dogs per capita 1000'] = percapita_df['dog_pop'] / percapita_df['population']*1000
```

```python
percapita_df
```
