## Grain Tariffs 2013

Processing/Cleaning: reading pdf using Tabula PY and converting into a dataframe

Source: http://open.canada.ca/data/en/dataset/a3ac17d5-2181-4360-accf-6a85fd3abc29

```python
import pandas as pd
%matplotlib inline
import numpy as np
import tabula
```

```python
#encoding issues needed to be fixed
df = pd.read_csv("tabula-tariff_2013_west.csv", encoding='ISO-8859-1')
df[:2]
```

```python
df = df.rename(columns= {"Unnamed: 0": "company"})
df
```

```python
#Which type of grain received the highest tariffs?
#Work on regular expressions!!!
df["Oats"].str.contains("/d.*/+", na=False, case=True, regex=True)
```

```python
df.loc[10] 
```

```python
df.loc[1]
```

```python
df.drop(10, inplace=True)
df.drop(1, inplace=True)
```

```python
df
```

```python
df.dtypes
```

```python
df.replace('-' ,np.nan, inplace=True)
```

```python
df
```

```python
df["Wheat incl durum"] = df["Wheat incl durum"].astype(float)
df["Oats"] = df["Oats"].astype(float)
df["Barley"] = df["Barley"].astype(float)
df["Rye"] = df["Rye"].astype(float)
df["Flaxseed"] = df["Flaxseed"].astype(float)
df["Canola, Rapeseed, or Mustard"] = df["Canola, Rapeseed, or Mustard"].astype(float)
df["Corn"] = df["Corn"].astype(float)
df["Sunflower Seed"] = df["Sunflower Seed"].astype(float)
df["Soybeans or Peas"] = df["Soybeans or Peas"].astype(float)
df["Other Grains"] = df["Other Grains"].astype(float)
```

```python
df.dtypes
```

```python
#Which grain has the highest tarrifs?
#Our data is wide and we need to melt it into a long df

long_tariffs = df.melt(value_vars=[
"Wheat incl durum",
"Oats",
"Barley",
"Rye",
"Flaxseed",
"Canola, Rapeseed, or Mustard",
"Corn",
"Sunflower Seed",
"Soybeans or Peas",
"Other Grains"
])
```

```python
long_tariffs.head()
```

```python
long_tariffs = long_tariffs.rename(columns= {"variable": "grain"})
long_tariffs.head()
```

```python
long_tariffs.groupby("grain").sum()
```

```python
#Let's graph it
#Oats have the greatest tariffs imposed and corn has the least
long_tariffs.groupby("grain").sum().sort_values(by="value", ascending=True).plot(kind='barh')
```

```python
#Which company has the highest tariffs imposed on them?
df.groupby("company").sum()
```

```python
#Hudson Bay Port Co. Churchill and Kinder Morgan have the greatest tariffs imposed
#Parrish & Heimbecker ÛÒ B.C. has the least tariffs imposed which makes sense because they mainly store corn and soybeans
df.sum(axis=1).sort_values(ascending=False)
```
